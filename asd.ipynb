```python
# Import necessary libraries
import pandas as pd

# Define the path to the CSV file
csv_file_path = "results/official_results_all.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# List of columns to calculate the average for
columns_to_average = ['MSRpar', 'MSRvid', 'SMT-eur', 'On-WN', 'SMT-news']

# Filter rows where 'Rank_ALL' is between 1 and 10 (inclusive)
filtered_df = df[(df['Rank_ALL'] >= 1) & (df['Rank_ALL'] <= 10)]

# Check if there are any rows after filtering
if not filtered_df.empty:
    # Calculate the average for each specified column in the filtered DataFrame
    averages_filtered = filtered_df[columns_to_average].mean()
    
    # Display the calculated averages for the filtered DataFrame
    print("=== Average Values for Rank_ALL between 1 and 10 ===")
    for column, avg in averages_filtered.items():
        print(f"{column}: {avg:.4f}")
else:
    print("No rows found with Rank_ALL between 1 and 10. Averages cannot be computed.")
```
